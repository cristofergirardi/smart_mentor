```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
metrics_df = pd.read_csv(f"/content/drive/MyDrive/Papers - Mestrado Cristofer/Cristofer-Tópicos Avançados em Sistemas/Programas/Metrics_hypotheses.csv")
# metrics_h9_df = pd.read_csv(f"/content/drive/MyDrive/Papers - Mestrado Cristofer/Cristofer-Tópicos Avançados em Sistemas/Programas/Metrics_median_hypothese_h9.csv")
```

```python
metrics_df_bert = metrics_df.where(metrics_df['metric'].str.contains('bert')).dropna(how='all')
# metrics_df_bert_h9 = metrics_h9_df.where(metrics_df['metric'].str.contains('bert')).dropna(how='all')
```

```python
metrics_df_bert.where(metrics_df['model'].str.contains('ope')).dropna(how='all')
```

```python
# metrics_df_bert_h9.where(metrics_h9_df['model'].str.contains('llama')).dropna(how='all')
```

```python
metrics_df_bert['h'] = metrics_df_bert['h'].str.replace('h', 's')
```

```python
metrics_df_bert
```

```python
# Create a single plot
plt.figure(figsize=(10, 6))

column_model = 'h'
column_metric = 'similarity'
#
plt.plot(metrics_df[column_model], metrics_df[column_metric], marker='o', label='bert', linestyle='-', color='b')
plt.plot(metrics_df[column_model], metrics_df[column_metric], marker='o', label='codet5', linestyle='-', color='g')

# Customize plot
plt.title(f'OpenAI Performance Metrics ', fontsize=16)
plt.xlabel('Hypotheses', fontsize=12)
plt.ylabel(f'{column_metric}', fontsize=12)
plt.legend(title='Metrics', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Show plot
plt.show()
```
